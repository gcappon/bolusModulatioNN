```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cross_validation import train_test_split

from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
```

```python
DATAFILE = os.path.join('data','data.csv')
TARGETFILE = os.path.join('data','target.csv')
OUTDIR = os.path.join('results')
```

```python
train = pd.read_table(DATAFILE,sep=',')

target = pd.read_table(TARGETFILE,sep=',')

dataset = pd.concat([train,target],axis=1)
```

```python
dataset.describe()
```

```python
dataset['Y'].describe()
```

```python
sns.distplot(dataset['Y'])
plt.show()
```

```python
#correlation matrix
corrmat = dataset.corr()
sns.heatmap(corrmat, vmax=.8, square=True)
plt.show()
```

```python
#Target correlation matrix
k = 5 #number of variables for heatmap
cols = corrmat.nlargest(k, 'Y')['Y'].index
cm = np.corrcoef(dataset[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

```python
#scatterplot
sns.set()
sns.pairplot(dataset, size = 2.5)
plt.show();
```

# Variable transformation

(Salta G_c, ROC, var_class (categoriche))

```python
#u2ss
var = 'u2ss'
sns.distplot(dataset[var])
plt.show()
res = stats.probplot(dataset[var], plot=plt)
plt.show()

a = np.log(dataset[var])
scaler = StandardScaler().fit(a)
a = scaler.fit_transform(a)
sns.distplot(a)
plt.show()
res = stats.probplot(a, plot=plt)
plt.show()
```

```python
#BW
var = 'BW'
sns.distplot(dataset[var])
plt.show()
res = stats.probplot(dataset[var], plot=plt)
plt.show()

#log-transformation and standard scaling
a = np.log(dataset[var])
scaler = StandardScaler().fit(a)
a = scaler.fit_transform(a)
sns.distplot(a)
plt.show()
res = stats.probplot(a, plot=plt)
plt.show()
```
